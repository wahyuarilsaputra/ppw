```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
}
#Cari Berdasarkan
pencarian = 'Prodi'
kode = 1
pta_list = []
for halaman in range(1,10):
  if pencarian == 'Fakultas':
    url = f"https://pta.trunojoyo.ac.id/c_search/byfac/{kode}/{str(halaman+1)}"
  elif pencarian == 'Prodi':
    url = f"https://pta.trunojoyo.ac.id/c_search/byprod/{kode}/{str(halaman+1)}"
  req = requests.get(url)
  soup = BeautifulSoup(req.text, 'lxml')
  prodi = soup.find('h2').text.replace('Journal','')
  list_ = soup.find('div', {'id':'content_journal'})

  for pta in list_.findAll('li',{'data-cat':'#luxury'}):
    span_elements = pta.find_all('span')
    Nama = span_elements[0].get_text(strip=True).replace('Penulis : ', '')
    pembimbing1 = span_elements[1].get_text(strip=True).replace('Dosen Pembimbing I : ', '')
    pembimbing2 = span_elements[2].get_text(strip=True).replace('Dosen Pembimbing II :', '')
    # pembimbing2 =
    link = pta.find('a', 'gray button')['href']
    req = requests.get(link)
    soup2 = BeautifulSoup(req.text, 'lxml')
    judul = pta.a.text.strip('\n').replace('\r','').replace('\n','')
    abstrak = soup2.find('p', {'align':'justify'}).text.replace('ABSTRAK','').replace('\r','').replace('\n','').replace('\t','')

    pta_list.append([judul,Nama,pembimbing1,pembimbing2,abstrak,prodi])
```

```python
frame_pta = pd.DataFrame(pta_list, columns=['Judul','Nama Penulis','Pembimbing I','Pembimbing II','Abstrak',pencarian])
frame_pta
```
